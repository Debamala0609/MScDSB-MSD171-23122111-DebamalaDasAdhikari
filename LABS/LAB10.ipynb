```python
class StudentManager:
    def __init__(self):
        self.students = {}

    def add_student(self, roll_number, name, marks):
        self.students[roll_number] = {"name": name, "marks": marks}

    def view_student(self, roll_number):
        student = self.students.get(roll_number)
        if student:
            print(f"Roll Number: {roll_number}")
            print(f"Name: {student['name']}")
            print(f"Marks: {student['marks']}")
        else:
            print("Student not found.")

    def modify_student(self, roll_number, name, marks):
        if roll_number in self.students:
            self.students[roll_number]["name"] = name
            self.students[roll_number]["marks"] = marks
            print(f"Student details updated successfully.")
        else:
            print("Student not found.")

def main():
    student_manager = StudentManager()

    while True:
        print("=" * 40)
        print("Student Management System")
        print("=" * 40)
        print("1. Add Student")
        print("2. View Student")
        print("3. Modify Student")
        print("4. Exit")
        print("=" * 40)

        choice = input("Please Enter Your choice: ")

        if choice == "1":
            roll_number = input("Enter Roll Number: ")
            name = input("Enter Name: ")
            marks = input("Enter Marks: ")
            student_manager.add_student(roll_number, name, marks)
            print("Student added successfully.")
        elif choice == "2":
            roll_number = input("Enter Roll Number: ")
            student_manager.view_student(roll_number)
        elif choice == "3":
            roll_number = input("Enter Roll Number: ")
            if roll_number in student_manager.students:
                name = input("Enter New Name: ")
                marks = input("Enter New Marks: ")
                student_manager.modify_student(roll_number, name, marks)
            else:
                print("Student not found.")
        elif choice == "4":
            print("Exiting the program")
            break
        else:
            print("Invalid Input")

if __name__ == "__main__":
    main()
```
